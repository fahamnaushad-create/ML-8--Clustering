##1. What is unsupervised learning in the context of machine learning?
- Answer

  Unsupervised learning is a type of machine learning where the model is trained on data without labels. The goal is to find hidden patterns, groupings, or structures in the data. Clustering (like K-Means) and dimensionality reduction (like PCA) are common unsupervised methods. It is useful when we don’t know the correct output beforehand.

##2. How does K-Means clustering algorithm work?
- Answer

  K-Means groups data into K clusters by minimizing the distance between points and their assigned cluster center. It starts with K random centroids, assigns each point to the nearest centroid, then recomputes centroids as the mean of assigned points. These two steps (assign + update) repeat until centroids stop changing much.

##3. Explain the concept of a dendrogram in hierarchical clustering?
- Answer

  A dendrogram is a tree-like diagram that shows how clusters are merged or split step by step in hierarchical clustering. Each join in the tree represents when two clusters combined. The height of a join shows how different those clusters were. By cutting the dendrogram at a chosen height, we can choose a desired number of clusters.

##4. What is the main difference between K-Means and Hierarchical Clustering?
- Answer

  K-Means requires you to predefine K (number of clusters) and works by iteratively updating centroids, producing one final partition. Hierarchical clustering builds a tree of clusters (bottom-up or top-down) and doesn’t require K upfront; you can choose K later by cutting the dendrogram. K-Means is faster on large datasets, while hierarchical gives more structure.

##5. What are the advantages of DBSCAN over K-Means?
- Answer

  DBSCAN can find clusters of arbitrary shapes and automatically detects noise/outliers, while K-Means assumes spherical clusters. DBSCAN doesn’t need the number of clusters K in advance. It also works better when clusters have different sizes and densities, where K-Means often fails.

##6. When would you use Silhouette Score in clustering?
- Answer

  Silhouette Score is used to evaluate how well data points fit within their assigned clusters compared to other clusters. It’s useful when deciding the best number of clusters (K). A higher score means better-defined clusters. It helps compare different clustering solutions without needing ground truth labels.

##7. What are the limitations of Hierarchical Clustering?
- Answer

  Hierarchical clustering can be computationally expensive for large datasets (O(n²) memory/time). Once clusters are merged or split, the decisions cannot be undone. It is also sensitive to noise and outliers, and results can vary with choice of distance metric and linkage method.

##8. Why is feature scaling important in clustering algorithms like K-Means?
- Answer

  K-Means uses distance (usually Euclidean) to assign points to clusters. Features with larger scales dominate the distance calculation. Scaling (e.g. StandardScaler or MinMaxScaler) puts all features on a similar scale so each variable contributes fairly, leading to more meaningful clusters.

##9. How does DBSCAN identify noise points?
- Answer

  DBSCAN defines clusters based on density. A point is considered a core point if it has enough neighbors within a given radius. Points that are not core points and not reachable from any core point are labeled as noise. These noise points do not belong to any cluster.

##10. Define inertia in the context of K-Means.
- Answer

  Inertia is the sum of squared distances between each data point and its assigned cluster center. It measures how compact the clusters are. Lower inertia generally means tighter clusters, but very low inertia can also mean overfitting if K is too large.

##11. What is the elbow method in K-Means clustering?
- Answer

  The elbow method helps choose a good value for K by plotting inertia vs. K. As K increases, inertia decreases. At some point, the improvement slows and the curve forms an “elbow.” That K value is often a good trade-off between complexity and performance.

##12. Describe the concept of "density" in DBSCAN.
- Answer

  In DBSCAN, density refers to how many points lie within a certain distance (eps) around a point. A region is considered dense if a point has at least min_samples neighbors within eps. Dense regions form clusters, while sparse areas are noise.

##13. Can hierarchical clustering be used on categorical data?
- Answer

  Yes, but not directly with Euclidean distance. You need appropriate distance measures (like Hamming distance) or first convert categories to numeric representations. Some specialized methods and similarity matrices can handle categorical data, but standard hierarchical clustering works best on numeric features.

##14. What does a negative Silhouette Score indicate?
- Answer

  A negative Silhouette Score means that many samples are closer (on average) to another cluster than to their own. This suggests poor clustering: points may be misassigned or clusters overlap heavily. It usually indicates that the chosen number of clusters or method is not suitable.

##15. Explain the term "linkage criteria" in hierarchical clustering.
- Answer

  Linkage criteria define how the distance between two clusters is calculated. Common linkages include single (minimum distance between points), complete (maximum distance), average (average distance), and ward (minimize variance). The choice of linkage can greatly affect the cluster shapes and the dendrogram.

##16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?
- Answer

  K-Means assumes clusters are roughly equal-sized, spherical, and have similar densities. When clusters differ in size or density, K-Means may pull centroids toward dense or large groups, misclassifying smaller or sparse clusters. It also struggles with non-spherical clusters.

##17. What are the core parameters in DBSCAN, and how do they influence clustering?
- Answer

  The two main parameters are eps (radius) and min_samples (minimum neighbors for a core point).

Larger eps means bigger neighborhoods and fewer, larger clusters.

Higher min_samples requires more neighbors to form a core; fewer points will be labeled as core, possibly more noise.
Choosing them correctly is critical to good clustering.

##18. How does K-Means++ improve upon standard K-Means initialization?
- Answer

  K-Means++ chooses initial centroids more intelligently instead of random selection. The first centroid is chosen randomly, and subsequent centroids are chosen with probability proportional to their distance from existing centroids. This spreads out initial centers and reduces the chance of poor convergence and bad local minima.

##19. What is agglomerative clustering?
- Answer

  Agglomerative clustering is a type of bottom-up hierarchical clustering. Each data point starts as its own cluster, and the algorithm repeatedly merges the two closest clusters based on a distance metric and linkage criteria. This continues until only one cluster remains or a stopping condition is met.

##20. What makes Silhouette Score a better metric than just inertia for model evaluation**
- Answer

  Inertia only measures cluster compactness and always decreases as K increases, which can be misleading. Silhouette Score considers both cohesion (within-cluster similarity) and separation (distance to other clusters). It provides a normalized score between -1 and 1, making it easier to compare different K values and clustering methods.

##1. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot.

###✅Python Code:

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X, y_true = make_blobs(n_samples=400, centers=4, cluster_std=1.0, random_state=42)

kmeans = KMeans(n_clusters=4, random_state=42)
labels = kmeans.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=30)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='red', marker='X', s=200, label='Centers')
plt.title("Q1: K-Means on make_blobs (4 centers)")
plt.legend()
plt.show()
```

##11. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries.

###✅Python Code:

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

X, _ = make_blobs(n_samples=400, centers=3, cluster_std=2.0, random_state=42)

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Create grid
x_min, x_max = X[:,0].min()-1, X[:,0].max()+1
y_min, y_max = X[:,1].min()-1, X[:,1].max()+1
xx, yy = np.meshgrid(np.linspace(x_min,x_max,300),
                     np.linspace(y_min,y_max,300))
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3, cmap='viridis')
plt.scatter(X[:,0], X[:,1], c=kmeans.labels_, cmap='viridis', s=30)
plt.title("Q11: K-Means with decision boundaries")
plt.show()
```

##14. Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a line plot.

##28. Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the clusters



###✅Python Code:


```python
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

digits = load_digits()
X = digits.data

tsne = TSNE(n_components=2, random_state=42, init='pca', learning_rate='auto')
X_2d = tsne.fit_transform(X)

agg = AgglomerativeClustering(n_clusters=10, linkage='ward')
labels = agg.fit_predict(X_2d)

plt.scatter(X_2d[:,0], X_2d[:,1], c=labels, cmap='tab10', s=5)
plt.title("Q28: Agglomerative Clustering on t-SNE reduced Digits")
plt.show()
```
